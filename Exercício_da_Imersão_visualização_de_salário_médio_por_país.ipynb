```python
# Importação das bibliotecas necessárias
import plotly.express as px
import pandas as pd
```

```python
# Criação de data frame com o endereço do arquivo
df = pd.read_csv("https://raw.githubusercontent.com/guilhermeonrails/data-jobs/refs/heads/main/salaries.csv")
```

```python
# Impressão de parte dos dados do arquivo
df.head()
```

```python
# Tradução dos nomes das colunas para o português
column_translations = {
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'empresa',
    'company_size': 'tamanho_empresa'
}
df = df.rename(columns=column_translations)
```

```python
# Eliminação de registros nulos
df_limpo = df.dropna()

# Contagem de quantidade de colunas do data frame com valores nulos
df_limpo.isnull().sum()
```
